```python
from nltk.corpus import stopwords
stopwords.words("english")[:]
```

```python
#function to remove stopwords and covert text into lowercase

def rm(sms):
    sms_no_punc = [ch for ch in sms if ch not in string.punctuation]
    sms_no_punc = "".join(sms_no_punc).split()
    
    sms_no_punc_no_stop = [word.lower() for word in sms_no_punc if word.lower() not in stopwords.words("english")]
    return sms_no_punc_no_stop
```

```python
df['text'].apply(rm).head(10)
```

```python
df["text"].apply(rm).head(15)
```

```python
def stemmer(text):
    text = text.split()
    words = ""
    for i in text:
        stemmer = PorterStemmer()
        words  +=(stemmer.stem(i))+ " "
    return words
```

```python
df["text"].apply(stemmer).head(10)
```

```python
def rm_punc(text):
    new_text = ''.join([char for char in text if char not in string.punctuation])
    return new_text
```

```python
df['new_text']=df['text'].apply(lambda row : rm_punc(row))
```
